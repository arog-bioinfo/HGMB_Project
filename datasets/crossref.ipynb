```python
import pandas as pd
import numpy as np
```

Load dataset

```python
df_v3 = pd.read_csv("/home/arog/Documents/GitHub/HGMB_Project/datasets/GFKB/GutFeelingKnowledgeBase-v4-Master_List.csv")
df_v5 = pd.read_csv("/home/arog/Documents/GitHub/HGMB_Project/datasets/GFKB/GFKB_v5_PreDiabetes.csv")
```

```python
df_v3.dropna(how='all', inplace=True)
df_v3.head()
```

```python
df_v5.dropna(how='all', inplace=True)
df_v5.head()
```

Prepare dataset for crossref with only healthy gut specimens

```python
# Select only "NCBI taxonomy name" and add new columns
filtered_df_v3 = df_v3[['UP name','Assembly ID (UP matched)','Present in GFKB v3 (Y/N)','Present in GFKB_epilepsy v3 (Y/N)']].copy()
filtered_df_v5 = df_v5[['UP name','Assembly ID (UP matched)','Present in GFKB v3 (Y/N)','Present in GFKB_epilepsy v3 (Y/N)']].copy()

# Prepare for concatenation
filtered_df_v3['Present in GFKB v5 (Y/N)'] = 'N'
filtered_df_v5['Present in GFKB v5 (Y/N)'] = 'Y'

# Concatenation
# Concatenate along rows (axis=0)
result_df = pd.concat([filtered_df_v3, filtered_df_v5], ignore_index=True)

# Add new columns (default 'N' and 'NA')
result_df['In BiGG Models (Y/N)'] = 'N'  # Default 'N' (updated later in crossref)
result_df['BiGG Model Link'] = 'NA'      # Default 'NA' (can be updated later in crossref)

# Sort by Name for easier crossreferecing
result_df = result_df.sort_values(by=['UP name'])

#Remove redundant rows
result_df = result_df.drop_duplicates(subset=['UP name','Present in GFKB v3 (Y/N)','Present in GFKB_epilepsy v3 (Y/N)']) #removes completely equal rows
duplicates = result_df[result_df['UP name'].duplicated(keep=False)]
```
