### Load dataset

```python
df_v3 = pd.read_csv("/home/arog/Documents/GitHub/HGMB_Project/datasets/GFKB/GutFeelingKnowledgeBase-v4-Master_List.csv")
df_v3_abund = pd.read_csv(
    "/home/arog/Documents/GitHub/HGMB_Project/datasets/GFKB/GutFeelingKnowledgeBase-v4-Epilepsy_Data.csv",
    skiprows=1
)
df_v5 = pd.read_csv("/home/arog/Documents/GitHub/HGMB_Project/datasets/GFKB/GFKB_v5_PreDiabetes.csv")
```

```python
df_v3.dropna(how='all', inplace=True)
df_v3.head()
```

```python
df_v5.dropna(how='all', inplace=True)
df_v5 = df_v5.dropna(axis=1, how='all')
df_v5.head()
```

```python
df_v3_abund.dropna(how='all', inplace=True)
df_v3_abund.head()
```

```python
def summarize_df(name, df, key_cols=5):
    total_cells = df.size
    missing_cells = df.isna().sum().sum()
    missing_pct = round((missing_cells / total_cells) * 100, 2)

    summary = {
        "Dataset": name,
        "Shape": f"{df.shape[0]} rows × {df.shape[1]} cols",
        "Columns": ", ".join(df.columns[:key_cols]) + ("..." if len(df.columns) > key_cols else ""),
        "Missing % (total)": f"{missing_pct}%"
    }
    return summary


# Collect summaries
summaries = [
    summarize_df("GFKB v4 Master List", df_v3),
    summarize_df("GFKB v4 Epilepsy Abund", df_v3_abund),
    summarize_df("GFKB v5 PreDiabetes", df_v5)
]

# Convert to DataFrame and display
summary_df = pd.DataFrame(summaries)
summary_df
```

### Prepare dataset for crossref

```python
# Select only "NCBI taxonomy name" and add new columns
filtered_df_v3 = df_v3[['Unique ID','UP name','NCBI taxonomy name','Assembly ID (UP matched)','Present in GFKB v3 (Y/N)','Present in GFKB_epilepsy v3 (Y/N)']].copy()
filtered_df_v5 = df_v5[['Unique ID','UP name','NCBI taxonomy name','Assembly ID (UP matched)','Present in GFKB v3 (Y/N)','Present in GFKB_epilepsy v3 (Y/N)']].copy()

# Prepare for concatenation
filtered_df_v3['Present in GFKB v5 (Y/N)'] = 'N'
filtered_df_v5['Present in GFKB v5 (Y/N)'] = 'Y'

# Concatenation
# Concatenate along rows (axis=0)
result_df = pd.concat([filtered_df_v3, filtered_df_v5], ignore_index=True)

# Add new columns (default 'N' and 'NA')
result_df['In BiGG Models (Y/N)'] = 'Y/N'  # Default 'Y/N' (updated later in crossref)
result_df['BiGG Model Link'] = None      # Default 'None' (can be updated later in crossref)

# Sort by Name for easier crossreferecing
result_df = result_df.sort_values(by=['UP name'])

#Remove redundant rows
result_df = result_df.drop_duplicates(subset=['UP name','NCBI taxonomy name','Present in GFKB v3 (Y/N)','Present in GFKB_epilepsy v3 (Y/N)']) #removes completely equal rows
duplicates = result_df[result_df['UP name'].duplicated(keep=False)]
```
